```python
import bizlens as bl
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# bl.set_profiling(True)
```

## 1. Load Titanic Dataset

```python
df = bl.load_dataset("titanic")
print(f"Dataset shape: {df.shape}")
bl.describe(df)
```

## 2. Data Quality Check

```python
bl.quality.data_profile(df)
bl.quality.completeness_report(df)
```

## 3. Contingency Tables with BizLens

```python
print("=== Gender vs Survival ===")
table, stats = bl.tables.contingency_table(df, 'sex', 'survived')
```

```python
print("=== Passenger Class vs Survival ===")
table, stats = bl.tables.contingency_table(df, 'pclass', 'survived')
```

## 4. Chi-Square Test of Independence

```python
# Using BizLens built-in contingency with chi-square results
table, chi_results = bl.tables.contingency_table(df, 'sex', 'survived')
print("Chi-Square Test Results:")
print(f"Chi² = {chi_results['chi2']:.4f}")
print(f"p-value = {chi_results['p_value']:.4f}")
print(f"Cramér\'s V = {chi_results['cramers_v']:.4f}")
print(chi_results['interpretation'])
```

## 5. Visual Association Analysis

```python
# Mosaic plot using seaborn
sns.catplot(x='sex', hue='survived', col='pclass', data=df, kind='count', height=4, aspect=0.8)
plt.suptitle('Survival by Gender and Passenger Class', y=1.02)
plt.show()
```

## 6. Probability & Expected Frequencies

```python
contingency = pd.crosstab(df['sex'], df['survived'], margins=True)
print("Observed Frequencies:")
print(contingency)

# Expected frequencies under independence
row_totals = contingency.iloc[:-1, -1]
col_totals = contingency.iloc[-1, :-1]
grand_total = contingency.iloc[-1, -1]
expected = np.outer(row_totals, col_totals) / grand_total

print("\nExpected Frequencies (if independent):")
print(pd.DataFrame(expected, index=row_totals.index, columns=col_totals.index))
```
