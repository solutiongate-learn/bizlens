# Process Mining with BizLensAnalyze **business processes, workflows, and event sequences** using BizLens process mining module.Key Concepts:- **Event Log**: Sequence of activities (case_id, activity, timestamp)- **Case Metrics**: Duration, activity count per case- **Bottlenecks**: Activities with longest processing times- **Variants**: Different sequences through the process- **Rework**: Cases that repeat activities

```python
import subprocess, syssubprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "bizlens", "plotly"])import bizlens as bzimport pandas as pdimport numpy as npfrom datetime import datetime, timedelta
```

## Dataset: Customer Support Process

```python
# Create event log for support ticket processnp.random.seed(42)events = []base_time = datetime(2024, 1, 1)for case_id in range(1, 101):  # 100 cases    activities = ['Create Ticket', 'Assign Agent', 'Investigate', 'Resolve', 'Close']    current_time = base_time + timedelta(days=case_id)    for activity in activities:        events.append({            'case_id': f'TICKET-{case_id:03d}',            'activity': activity,            'timestamp': current_time,            'agent': np.random.choice(['Alice', 'Bob', 'Charlie']),            'duration_hours': np.random.gamma(2, 2)        })        current_time += timedelta(hours=np.random.gamma(2, 2))event_log = pd.DataFrame(events)print(f"Event log created: {len(event_log)} events across {event_log['case_id'].nunique()} cases\n")print(event_log.head(10))
```

## 1. Case-Level Metrics

```python
# Get case metrics (duration, activity count, etc.)case_metrics = bz.process_mining.case_metrics(event_log)print("\n=== Case Metrics ===")print(case_metrics.head())print(f"\nAverage case duration: {case_metrics['case_duration'].mean():.1f} hours")print(f"Min/Max activities per case: {case_metrics['activity_count'].min()}-{case_metrics['activity_count'].max()}")
```

## 2. Activity Metrics

```python
# Analyze each activity's characteristicsactivity_metrics = bz.process_mining.activity_metrics(event_log)print("\n=== Activity Metrics ===")print(activity_metrics)
```

## 3. Resource (Agent) Utilization

```python
# Analyze workload by agentresource_analysis = bz.process_mining.resource_analysis(event_log)print("\n=== Resource Utilization ===")print(resource_analysis)
```

## 4. Process Variants

```python
# Discover different execution paths (variants)variants = bz.process_mining.variant_discovery(event_log)print("\n=== Process Variants ===")for i, (variant, count, freq) in enumerate(variants, 1):    print(f"Variant {i} ({count} cases, {freq:.1f}%): {' → '.join(variant)}")
```

## 5. Bottleneck Analysis

```python
# Identify bottlenecks (longest activities)bottlenecks = bz.process_mining.bottleneck_analysis(event_log)print("\n=== Process Bottlenecks ===")print(bottlenecks)print(f"\nMost time-consuming: {bottlenecks.iloc[0]['activity']} ({bottlenecks.iloc[0]['avg_duration']:.1f}h)")
```

## 6. Rework Detection

```python
# Find cases with repeated activities (rework)rework = bz.process_mining.rework_detection(event_log)print(f"\n=== Rework Analysis ===")print(f"Cases with rework: {len(rework)} out of {event_log['case_id'].nunique()}")if len(rework) > 0:    print("\nCases with repeated activities:")    print(rework)
```

## 7. Process Flow (Transition Matrix)

```python
# Generate transition matrix (which activities follow which)transitions = bz.process_mining.transition_matrix(event_log)print("\n=== Activity Transition Matrix ===")print(transitions)print("\nInterpretation: Matrix[i,j] = how often activity i is followed by activity j")
```

## Key Insights from Process Mining✓ **Process Flow**: Ticket → Agent → Investigate → Resolve → Close✓ **Bottleneck**: Investigation phase takes longest✓ **Variants**: Some cases skip investigation or have rework✓ **Resources**: Workload reasonably distributed among agents**Use Cases**:- Identify process inefficiencies- Optimize resource allocation- Detect process deviations- Plan process improvements
