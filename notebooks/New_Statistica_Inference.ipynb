## Scenario 1: A/B Testing (Two Independent Groups)

```python
np.random.seed(42)
control = np.random.normal(loc=100, scale=15, size=60)
treatment = np.random.normal(loc=108, scale=15, size=60)

print(f"Control   : mean = {control.mean():.2f}, n = {len(control)}")
print(f"Treatment : mean = {treatment.mean():.2f}, n = {len(treatment)}")
```

```python
# Confidence Intervals
bl.inference.confidence_interval(control, confidence=0.95)
bl.inference.confidence_interval(treatment, confidence=0.95)
```

```python
# Two-Sample t-test
result = bl.inference.two_sample_ttest(control, treatment)
print(f"p-value = {result['p_value']:.4f}")
print("Effect Size (Cohen's d):", result['cohens_d'])
```

## Scenario 2: Paired Before-After Study

```python
before = np.random.normal(75, 12, 50)
after = before + np.random.normal(8, 5, 50)   # treatment effect

paired_result = bl.inference.paired_ttest(before, after)
print(f"Mean difference = {paired_result['mean_difference']:.2f}, p-value = {paired_result['p_value']:.4f}")
```

## Scenario 3: One-Way ANOVA (3+ Groups)

```python
group_a = np.random.normal(100, 15, 40)
group_b = np.random.normal(105, 15, 40)
group_c = np.random.normal(118, 15, 40)

groups_dict = {'A': group_a, 'B': group_b, 'C': group_c}
anova = bl.inference.anova_test(groups_dict)
print(f"ANOVA p-value = {anova['p_value']:.4f} | Effect size (η²) = {anova['eta_squared']:.3f}")
```

## Scenario 4: Correlation Analysis

```python
x = np.random.normal(100, 20, 120)
y = x * 0.7 + np.random.normal(0, 15, 120)

corr = bl.inference.correlation_test(x, y)
print(f"Pearson r = {corr['pearson_r']:.3f}, p-value = {corr['pearson_p']:.4f}")
```

## Effect Size Interpretation (Educational)

```python
print("Effect Size Interpretation using BizLens:")
print(bl.inference.effect_size_interpreter(0.35, "cohens_d"))
print(bl.inference.effect_size_interpreter(0.08, "eta_squared"))
```

## Visual Summary

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

axes[0,0].hist(control, bins=15, alpha=0.7, label='Control', color='blue')
axes[0,0].hist(treatment, bins=15, alpha=0.7, label='Treatment', color='orange')
axes[0,0].set_title('A/B Test Distributions')
axes[0,0].legend()

axes[0,1].boxplot([before, after], labels=['Before', 'After'])
axes[0,1].set_title('Paired Before-After')

axes[1,0].scatter(x, y, alpha=0.6)
axes[1,0].set_title(f'Correlation (r = {corr["pearson_r"]:.3f})')

axes[1,1].hist(group_a, alpha=0.6, label='A', color='blue')
axes[1,1].hist(group_b, alpha=0.6, label='B', color='green')
axes[1,1].hist(group_c, alpha=0.6, label='C', color='red')
axes[1,1].set_title('Multiple Groups (ANOVA)')
axes[1,1].legend()

plt.tight_layout()
plt.show()
```
