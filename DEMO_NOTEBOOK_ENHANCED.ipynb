## 1️⃣ CENTRAL TENDENCY STATISTICS

### 📊 What is Central Tendency?

Central tendency measures where data "centers" around:

| Statistic | Formula | Meaning | When to Use |
|-----------|---------|---------|-------------|
| **Mean** | μ = Σx / n | Average | Symmetric distributions |
| **Median** | 50th percentile | Middle value | Skewed data, has outliers |
| **Mode** | Most frequent | Most common | Categorical or discrete data |

**Key Insight**: If Mean ≠ Median → Data is skewed!

```python
# Load dataset
df = bl.load_dataset('school_cafeteria')

print(f"Dataset: School Cafeteria Spending")
print(f"Size: {df.shape[0]} students × {df.shape[1]} columns\n")
print(df.head(10))
```

```python
# Create analyzer with color scheme
bd = bl.BizDesc(df, color_scheme='academic')  # 'academic', 'pastel', or 'vibrant'

# Get central tendency statistics (Mean, Median, Mode)
cent_tend = bd.central_tendency()
```

```python
# Interpret central tendency
print("""
🔍 INTERPRETATION OF RESULTS:
════════════════════════════════════════════════════════════

SPENDING Column:
  Mean (6.22) > Median (4.80)
  → Data is RIGHT-SKEWED (positive skew)
  → Most students spend little ($3-5)
  → Few students spend a LOT ($15+)
  → Mean is pulled up by outliers
  → MEDIAN is better for "typical" student

SATISFACTION Column:
  Mean ≈ Median
  → Data is roughly SYMMETRIC
  → More normal distribution
  → Mean and Median similar

Skewness Values:
  • < -0.5: Left-skewed (negative)
  • -0.5 to +0.5: Symmetric (normal-like)
  • > +0.5: Right-skewed (positive)
""")
```

---

## 4️⃣ VISUALIZATION 2: BOXPLOT (Quartiles + Outliers)

**Shows**: Q1, Median (Q2), Q3, Outliers
**IQR (Interquartile Range)**: Q3 - Q1 = Middle 50%
**Whiskers**: Valid range (without outliers)
**Dots**: Individual outliers

```python
bd.visualize('spending', plot_type='boxplot')
print("""
BOXPLOT ANATOMY:
  ├─ Bottom whisker: Min (without outliers)
  ├─ Bottom of box: Q1 (25th percentile)
  ├─ Line in box: MEDIAN (Q2, 50th percentile) ← Most important!
  ├─ Top of box: Q3 (75th percentile)
  ├─ Top whisker: Max (without outliers)
  └─ Dots: OUTLIERS (beyond 1.5 × IQR)

FORMULA FOR OUTLIERS:
  IQR = Q3 - Q1
  Lower bound = Q1 - 1.5 × IQR
  Upper bound = Q3 + 1.5 × IQR
  → Values outside bounds = outliers
""")
```

## 1️⃣2️⃣ OUTLIER DETECTION WITH VISUALIZATION

```python
# Detect outliers with boxplot visualization
outliers = bd_cafe.outliers(method='iqr')
print("""
OUTLIER DETECTION (IQR METHOD):

FORMULA:
  IQR = Q3 - Q1
  Lower Bound = Q1 - 1.5 × IQR
  Upper Bound = Q3 + 1.5 × IQR
  → Values outside bounds = OUTLIERS (shown as red dots)

WHY 1.5 × IQR?
  ✓ Statistical standard
  ✓ ~0.7% of normal data flagged as outliers
  ✓ Good balance between sensitivity & specificity

INTERPRETATION:
  ✓ Few outliers (< 5%) = Normal data
  ✓ Many outliers (> 10%) = Unusual data or errors
  ✓ Check outliers: Are they real or data entry errors?
""")
```

## 1️⃣3️⃣ NORMALITY TESTING (Shapiro-Wilk)

```python
# Test if data is normally distributed
normality = bd_cafe.normality_test()
print("""
SHAPIRO-WILK TEST:

NULL HYPOTHESIS (H0):
  "Data is normally distributed"

DECISION RULE:
  p-value > 0.05  →  FAIL to reject H0  →  Data likely NORMAL ✓
  p-value < 0.05  →  REJECT H0  →  Data likely NOT NORMAL ✗

RESULT FOR SPENDING:
  p-value should be < 0.05
  → Data is NOT normally distributed
  → Because it's RIGHT-SKEWED!
  → This is EXPECTED for real-world data

IMPORTANT:
  ✓ Real data is often NOT perfectly normal
  ✓ Slight deviations = Usually fine
  ✓ Highly skewed = May need transformation
  ✓ Always visualize BEFORE assuming normality
""")
```
