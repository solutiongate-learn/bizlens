## 3. Theoretical Probability Distributions

```python
# Normal Distribution (e.g. customer satisfaction)
normal_data = np.random.normal(loc=75, scale=12, size=1000)
bl.tables.percentile_table(pd.Series(normal_data))

plt.hist(normal_data, bins=50, alpha=0.7, color='skyblue', density=True)
plt.title('Normal Distribution - Simulated Customer Satisfaction')
plt.show()
```

```python
# Exponential Distribution (e.g. time between customer arrivals)
exp_data = np.random.exponential(scale=10, size=1000)
plt.hist(exp_data, bins=50, alpha=0.7, color='coral', density=True)
plt.title('Exponential Distribution - Inter-arrival Time')
plt.show()
```

```python
# Poisson Distribution (e.g. number of daily orders)
poisson_data = np.random.poisson(lam=45, size=1000)
bl.tables.frequency_table(pd.Series(poisson_data), top_n=15)

plt.hist(poisson_data, bins=30, alpha=0.7, color='green', density=True)
plt.title('Poisson Distribution - Daily Orders')
plt.show()
```

## 4. Monte Carlo Simulation (Business Example)

```python
# Simulate total revenue for 1000 days
n_sim = 10000
daily_orders = np.random.poisson(lam=45, size=n_sim)
avg_order_value = np.random.normal(45, 8, size=n_sim)

simulated_revenue = daily_orders * avg_order_value

print(f"Simulated Mean Daily Revenue: ${simulated_revenue.mean():,.2f}")
print(f"95% Confidence Interval: ${np.percentile(simulated_revenue, 2.5):,.2f} - ${np.percentile(simulated_revenue, 97.5):,.2f}")

plt.hist(simulated_revenue, bins=80, alpha=0.7, color='purple')
plt.title('Monte Carlo Simulation - Daily Revenue')
plt.xlabel('Revenue ($')
plt.show()
```
