# Descriptive Analytics with BizLens

This notebook covers **frequency tables, percentiles, contingency tables, and visualizations**.

```python
import subprocess, sys
subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "bizlens", "seaborn", "matplotlib"])

import bizlens as bz
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## Dataset: Customer Service Transactions

```python
np.random.seed(42)
n = 500

data = pd.DataFrame({
    'customer_segment': np.random.choice(['Premium', 'Standard', 'Basic'], n),
    'issue_category': np.random.choice(['Billing', 'Technical', 'Account', 'Product'], n),
    'resolution_time': np.random.gamma(shape=2, scale=10, size=n),
    'satisfaction': np.random.choice([1, 2, 3, 4, 5], n),
    'was_resolved': np.random.choice([True, False], n, p=[0.85, 0.15])
})

print(f"Dataset shape: {data.shape}")
print(data.head())
```

## 1. Frequency Tables

```python
freq_segment = bz.tables.frequency_table(data, 'customer_segment')
print("\n=== Customer Segment Distribution ===")
print(freq_segment)
```

## 2. Percentile Analysis

```python
percentiles = bz.tables.percentile_table(data[['resolution_time']])
print("\n=== Resolution Time Percentiles ===")
print(percentiles)
```

## 3. Contingency Tables

```python
contingency = bz.tables.contingency_table(
    data,
    row_col='customer_segment',
    col_col='issue_category'
)
print("\n=== Segment × Category ===")
print(contingency)
```

## 4. Data Quality

```python
quality = bz.quality.data_profile(data)
print(f"Quality Score: {quality:.1f}/100")

outliers = bz.diagnostic.detect_outliers(data[['resolution_time']], method='iqr')
print(f"Outliers: {len(outliers)} records")
```

## 5. Visualizations

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(data['resolution_time'], bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Resolution Time Distribution')
axes[0].set_xlabel('Hours')

satisfaction_freq = data['satisfaction'].value_counts().sort_index()
axes[1].bar(satisfaction_freq.index, satisfaction_freq.values, color='coral')
axes[1].set_title('Satisfaction Ratings')
axes[1].set_xlabel('Rating')

plt.tight_layout()
plt.show()
```
