# Process Mining with BizLens

Analyze **business processes, workflows, and event sequences**.

```python
import subprocess, sys
subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "bizlens", "plotly"])

import bizlens as bz
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

## Dataset: Support Ticket Workflow

```python
np.random.seed(42)
events = []
base_time = datetime(2024, 1, 1)

for case_id in range(1, 101):
    activities = ['Create', 'Assign', 'Investigate', 'Resolve', 'Close']
    current_time = base_time + timedelta(days=case_id)
    for activity in activities:
        events.append({
            'case_id': f'TICKET-{case_id:03d}',
            'activity': activity,
            'timestamp': current_time,
            'agent': np.random.choice(['Alice', 'Bob', 'Charlie'])
        })
        current_time += timedelta(hours=np.random.gamma(2, 2))

event_log = pd.DataFrame(events)
print(f"Events: {len(event_log)}, Cases: {event_log['case_id'].nunique()}")
```

## 1. Case Metrics

```python
case_metrics = bz.process_mining.case_metrics(event_log)
print("\n=== Case Metrics ===")
print(case_metrics.head())
```

## 2. Activity Analysis

```python
activity_metrics = bz.process_mining.activity_metrics(event_log)
print("\n=== Activity Metrics ===")
print(activity_metrics)
```

## 3. Bottleneck Analysis

```python
bottlenecks = bz.process_mining.bottleneck_analysis(event_log)
print("\n=== Bottlenecks ===")
print(bottlenecks)
```

## 4. Process Variants

```python
variants = bz.process_mining.variant_discovery(event_log)
print(f"\nFound {len(variants)} variants")
```

## 5. Resource Analysis

```python
resources = bz.process_mining.resource_analysis(event_log)
print("\n=== Agent Workload ===")
print(resources)
```
