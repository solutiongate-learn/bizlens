```python
import bizlens as bl
import pandas as pd
import numpy as np

# Optional: Enable performance timing (new feature in v2.2.13)
# bl.set_profiling(True)

np.random.seed(42)
data = pd.DataFrame({
    'product': np.random.choice(['A', 'B', 'C'], 100),
    'sales': np.random.exponential(1000, 100),
    'region': np.random.choice(['North', 'South', 'East', 'West'], 100),
    'quarter': np.random.choice(['Q1', 'Q2', 'Q3', 'Q4'], 100)
})

print(f"Created dataset with {len(data)} records")
print(data.head())

# ── Optional: Use a BizLens built-in dataset ─────────────────────────────
# df = bl.load_dataset('tips')  # restaurant tips — total_bill, tip, sex, smoker, day, time, size
# df = bl.load_dataset('titanic')  # passenger data — survived, pclass, sex, age, fare
#
# After loading, explore with:
# print(df.shape)         # rows × columns
# print(df.dtypes)        # column types
# print(df.head())        # first 5 rows
#
# ── Export / Save Results ─────────────────────────────────────────────────
# df.to_csv('output.csv', index=False)  # Save results to CSV
# df.to_excel('output.xlsx', index=False)  # Save as Excel
# df.to_json('output.json', orient='records', indent=2)  # Save as JSON
# ── Polars alternative (optional — for users who prefer polars) ──────────────
# import polars as pl
# df_pl = pl.from_pandas(bl.load_dataset('titanic'))    # load → convert to polars
# df_pl = pl.read_csv('your_data.csv')                  # or read directly
# df_pl = pl.DataFrame(data)                             # or create from dict
#
# BizLens accepts both pandas and polars DataFrames transparently:
# bl.tables.frequency_table(df_pl['sex'])               # works with polars too
# bl.quality.data_profile(df_pl)                        # works with polars too
#
# Convert polars back to pandas when needed:
# df = df_pl.to_pandas()
```

## Dataset Information & Rich Displays

```python
print("=== Basic Dataset Info ===")
print(f"Shape: {data.shape}")
print(f"Columns: {list(data.columns)}")
print(f"Missing values: {data.isnull().sum().sum()}")

# Functions print their output automatically - no need to print again
bl.quality.data_profile(data)
bl.quality.completeness_report(data)

# Sample data preview
print("\n=== Sample Data (first 8 rows) ===")
print(data.head(8))
```
