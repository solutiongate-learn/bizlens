# Descriptive Analytics with BizLensThis notebook covers **frequency tables, percentiles, contingency tables, and visualizations**.Perfect for:- Exploratory data analysis (EDA)- Creating professional statistical tables- Understanding data distributions- Identifying data quality issues

```python
import subprocess, syssubprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "bizlens", "seaborn", "matplotlib"])import bizlens as bzimport pandas as pdimport numpy as npimport matplotlib.pyplot as pltimport seaborn as snssns.set_style("whitegrid")plt.rcParams['figure.figsize'] = (12, 6)
```

## Dataset: Customer Service Transactions

```python
# Create realistic customer service datasetnp.random.seed(42)n = 500data = pd.DataFrame({    'customer_segment': np.random.choice(['Premium', 'Standard', 'Basic'], n),    'issue_category': np.random.choice(['Billing', 'Technical', 'Account', 'Product'], n),    'resolution_time': np.random.gamma(shape=2, scale=10, size=n),  # hours    'satisfaction': np.random.choice([1, 2, 3, 4, 5], n),    'was_resolved': np.random.choice([True, False], n, p=[0.85, 0.15])})print(f"Dataset shape: {data.shape}")print("\nFirst 5 rows:")print(data.head())
```

## 1. Frequency Tables

```python
# Frequency distribution by segmentfreq_segment = bz.tables.frequency_table(data, 'customer_segment')print("\n=== Customer Segment Distribution ===")print(freq_segment)
```

```python
# Frequency for issue categoriesfreq_issue = bz.tables.frequency_table(data, 'issue_category')print("\n=== Issue Category Distribution ===")print(freq_issue)
```

## 2. Percentile Analysis

```python
# Percentile analysis for resolution timepercentiles = bz.tables.percentile_table(data[['resolution_time']])print("\n=== Resolution Time Percentiles ===")print(percentiles)print(f"\nInsight: 50% of tickets resolved within {percentiles.loc[50].values[0]:.1f} hours")
```

## 3. Contingency Tables (Cross-tabulation)

```python
# Cross-tabulation: Segment vs Categorycontingency = bz.tables.contingency_table(    data,    row_col='customer_segment',    col_col='issue_category')print("\n=== Segment × Category Contingency Table ===")print(contingency)
```

## 4. Summary Statistics

```python
# Summary for numeric columnssummary = bz.tables.summary_statistics(data[['resolution_time', 'satisfaction']])print("\n=== Summary Statistics ===")print(summary)
```

## 5. Data Quality & Diagnostics

```python
# Completeness checkcompleteness = bz.quality.completeness_report(data)print("\n=== Data Completeness ===")print(completeness)# Overall quality scorequality = bz.quality.data_profile(data)print(f"\nOverall Quality Score: {quality:.1f}/100")
```

```python
# Outlier detection in resolution_timeoutliers_df = bz.diagnostic.detect_outliers(    data[['resolution_time']],    method='iqr')print(f"\n=== Outliers Detected: {len(outliers_df)} records ===")if len(outliers_df) > 0:    print(outliers_df[['resolution_time']].describe())
```

## 6. Visualizations

```python
# Distribution of resolution timefig, axes = plt.subplots(2, 2, figsize=(14, 10))# Histogramaxes[0,0].hist(data['resolution_time'], bins=30, color='skyblue', edgecolor='black')axes[0,0].set_title('Resolution Time Distribution', fontsize=12, fontweight='bold')axes[0,0].set_xlabel('Hours')axes[0,0].set_ylabel('Frequency')# Box plot by segmentdata.boxplot(column='resolution_time', by='customer_segment', ax=axes[0,1])axes[0,1].set_title('Resolution Time by Segment', fontsize=12, fontweight='bold')axes[0,1].set_xlabel('Segment')axes[0,1].set_ylabel('Hours')# Satisfaction distributionsatisfaction_freq = data['satisfaction'].value_counts().sort_index()axes[1,0].bar(satisfaction_freq.index, satisfaction_freq.values, color='coral', edgecolor='black')axes[1,0].set_title('Satisfaction Ratings', fontsize=12, fontweight='bold')axes[1,0].set_xlabel('Rating')axes[1,0].set_ylabel('Count')axes[1,0].set_xticks([1, 2, 3, 4, 5])# Correlation heatmapcorr = data[['resolution_time', 'satisfaction']].corr()sns.heatmap(corr, annot=True, cmap='coolwarm', ax=axes[1,1], vmin=-1, vmax=1)axes[1,1].set_title('Correlation Matrix', fontsize=12, fontweight='bold')plt.tight_layout()plt.show()
```
