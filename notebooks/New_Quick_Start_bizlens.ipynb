```python
import bizlens as bl
import pandas as pd
import numpy as np

# Optional: Enable performance timing (new feature in v2.2.12)
# bl.set_profiling(True)

np.random.seed(42)
data = pd.DataFrame({
    'product': np.random.choice(['A', 'B', 'C'], 100),
    'sales': np.random.exponential(1000, 100),
    'region': np.random.choice(['North', 'South', 'East', 'West'], 100),
    'quarter': np.random.choice(['Q1', 'Q2', 'Q3', 'Q4'], 100)
})

print(f"Created dataset with {len(data)} records")
print(data.head())
```

## Dataset Information & Rich Displays

```python
print("=== Basic Dataset Info ===")
print(f"Shape: {data.shape}")
print(f"Columns: {list(data.columns)}")
print(f"Missing values: {data.isnull().sum().sum()}")

# Functions print their output automatically - no need to print again
bl.quality.data_profile(data)
bl.quality.completeness_report(data)

# Sample data preview
print("\n=== Sample Data (first 8 rows) ===")
print(data.head(8))
```

## 1. Descriptive Analytics - Tables & Statistics

```python
# frequency_table prints automatically and returns RichTable object
# No need to print it again
bl.tables.frequency_table(data['product'])
```

```python
# summary_statistics prints automatically and returns RichTable object
# No need to print it again
bl.tables.summary_statistics(data)
```
