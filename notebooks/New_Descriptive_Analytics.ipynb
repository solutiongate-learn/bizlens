```python
import bizlens as bl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Enable performance timing
# bl.set_profiling(True)

np.random.seed(42)
n = 500

data = pd.DataFrame({
    'customer_segment': np.random.choice(['Premium', 'Standard', 'Basic'], n),
    'issue_category': np.random.choice(['Billing', 'Technical', 'Account', 'Product'], n),
    'resolution_time': np.random.gamma(shape=2, scale=10, size=n),
    'satisfaction': np.random.choice([1, 2, 3, 4, 5], n),
    'was_resolved': np.random.choice([True, False], n, p=[0.85, 0.15])
})

print(f"Dataset shape: {data.shape}")
print(data.head())
```

## 1. Frequency Tables

```python
freq_segment = bl.tables.frequency_table(data['customer_segment'])
freq_segment  # display rich table
```

## 2. Percentile Analysis

```python
percentiles = bl.tables.percentile_table(data['resolution_time'])
percentiles  # display rich table
```

## 3. Contingency Tables

```python
contingency_table, stats = bl.tables.contingency_table(data, 'customer_segment', 'issue_category')
contingency_table  # display rich table
```

## 4. Data Quality

```python
quality = bl.quality.data_profile(data)
print(f"Overall Quality Score: {quality['overall_score']:.1f}/100")
```

```python
completeness = bl.quality.completeness_report(data)
```

## 5. Visualizations

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(data['resolution_time'], bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Resolution Time Distribution')
axes[0].set_xlabel('Hours')

satisfaction_freq = data['satisfaction'].value_counts().sort_index()
axes[1].bar(satisfaction_freq.index, satisfaction_freq.values, color='coral')
axes[1].set_title('Satisfaction Ratings')
axes[1].set_xlabel('Rating')

plt.tight_layout()
plt.show()
```
