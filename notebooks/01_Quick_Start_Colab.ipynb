```python
# Step 2: Import and create sample dataimport bizlens as bzimport pandas as pdimport numpy as np# Create sample sales datanp.random.seed(42)data = pd.DataFrame({    'product': np.random.choice(['A', 'B', 'C'], 100),    'sales': np.random.exponential(1000, 100),    'region': np.random.choice(['North', 'South', 'East', 'West'], 100),    'quarter': np.random.choice(['Q1', 'Q2', 'Q3', 'Q4'], 100)})print(f"Created dataset with {len(data)} records")print(data.head())
```

## 1. Descriptive Analytics - Tables & Statistics

```python
# Generate frequency tablefrequency = bz.tables.frequency_table(data, 'product')print(frequency)
```

```python
# Summary statisticssummary = bz.tables.summary_statistics(data[['sales']])print(summary)
```

## 2. Data Quality Assessment

```python
# Quality profile (returns 0-100 score)quality_score = bz.quality.data_profile(data)print(f"Overall Data Quality Score: {quality_score:.1f}/100")# Completeness reportcompleteness = bz.quality.completeness_report(data)print(completeness)
```

## 3. Statistical Diagnostics

```python
# Detect outliers in salesoutliers = bz.diagnostic.detect_outliers(data[['sales']], method='iqr')print(f"Outliers detected: {len(outliers)} records")# Normality testnormality = bz.diagnostic.normality_test(data[['sales']])print(normality)
```

## 4. Statistical Inference

```python
# Confidence interval for mean salesci = bz.inference.confidence_interval(data['sales'], confidence=0.95)print(f"95% CI for mean sales: [{ci['lower']:.2f}, {ci['upper']:.2f}]")# Compare sales by region (ANOVA)anova_result = bz.inference.anova_test(    data,    group_col='region',    value_col='sales')print(f"ANOVA p-value: {anova_result['p_value']:.4f}")
```
