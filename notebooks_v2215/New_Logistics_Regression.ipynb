```python
import bizlens as bl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

sns.set_theme(style="whitegrid", palette="husl", font_scale=1.1)
```

## 1. Load and Prepare Data


```python
df = bl.load_dataset('titanic')
print(f"Dataset shape: {df.shape}")
df.head()
```

## 2. Data Exploration


```python
bl.describe(df)
bl.quality.data_profile(df)
```

## 3. Logistic Regression Model


```python
# Select features and target
df_model = df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'survived']].dropna()

# Encode categorical
X = pd.get_dummies(df_model[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']], drop_first=True)
X = X.astype(float)
y = df_model['survived'].astype(float)

X_const = sm.add_constant(X)

logit_model = sm.Logit(y, X_const).fit(disp=0)
print(logit_model.summary())
```
