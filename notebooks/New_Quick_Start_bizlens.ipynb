```python
import bizlens as bl
import pandas as pd
import numpy as np

# Optional: Enable performance timing (new feature in v2.2.12)
# bl.set_profiling(True)

np.random.seed(42)
data = pd.DataFrame({
    'product': np.random.choice(['A', 'B', 'C'], 100),
    'sales': np.random.exponential(1000, 100),
    'region': np.random.choice(['North', 'South', 'East', 'West'], 100),
    'quarter': np.random.choice(['Q1', 'Q2', 'Q3', 'Q4'], 100)
})

print(f"Created dataset with {len(data)} records")
print(data.head())
```

## Dataset Information & Rich Displays

```python
print("=== Basic Dataset Info ===")
print(f"Shape: {data.shape}")
print(f"Columns: {list(data.columns)}")
print(f"Missing values: {data.isnull().sum().sum()}")

# Rich quality report
bl.quality.data_profile(data)

# Completeness report
bl.quality.completeness_report(data)

# Sample data preview
print("\n=== Sample Data (first 8 rows) ===")
display(data.head(8))
```

## 1. Descriptive Analytics - Tables & Statistics

```python
freq = bl.tables.frequency_table(data['product'])
freq  # display the rich table
```

```python
summary = bl.tables.summary_statistics(data)
summary  # display the rich table
```

## 3. Statistical Diagnostics

```python
outliers_idx, outliers_table = bl.diagnostic.detect_outliers(data['sales'], method='iqr')
print(f"\nOutliers detected in sales: {len(outliers_idx)} records")
```

## 4. Statistical Inference Example

```python
lower, mean, upper = bl.inference.confidence_interval(data['sales'])
print(f"\n95% Confidence Interval for mean sales: [{lower:.2f}, {upper:.2f}]")
```
