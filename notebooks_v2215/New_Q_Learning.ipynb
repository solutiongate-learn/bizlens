```python
import seaborn as sns
import bizlens as bl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", palette="husl", font_scale=1.1)
```

## 1. Q-Learning Environment


```python
np.random.seed(42)

# Simulated Q-Learning rewards
episodes = 100
rewards = np.random.normal(loc=50, scale=15, size=episodes)

rewards_series = pd.Series(rewards)
print(f"Episodes: {len(rewards_series)}")
print(f"Mean Reward: {rewards_series.mean():.2f}")
print(f"Std Dev: {rewards_series.std():.2f}")
```

## 2. Reward Statistics


```python
bl.tables.summary_statistics(rewards_series.rename('Episode Reward').to_frame())
bl.tables.percentile_table(rewards_series.rename('Episode Reward'))
```

## 3. Reward Distribution


```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(rewards_series, bins=20, color='steelblue', edgecolor='black')
plt.title('Reward Distribution')
plt.xlabel('Reward')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.plot(rewards_series.values, marker='o', linestyle='-', color='steelblue')
plt.title('Rewards Over Episodes')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.tight_layout()
plt.show()
```
