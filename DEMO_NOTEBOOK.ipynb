## 2️⃣ Descriptive Statistics with Formulas

````python
# Initialize BizDesc
bd = bl.BizDesc(df)

# Get complete statistics
stats = bd.describe(include_plots=True, bins=30)
````

### 📐 What These Formulas Mean

**Mean (Average)**
```
μ = (Σ x) / n
```
Sum all values, divide by count. The "typical" value.

**Median (Middle)**
Sort data → take middle value. Better for skewed data than mean.

**Standard Deviation (Spread)**
```
σ = √(Σ(x - μ)² / (n-1))
```
How far values typically deviate from mean. Higher = more spread.

**Skewness (Asymmetry)**
```
γ = (3(μ - median)) / σ
```
- Positive: Right-skewed (long tail right) → like spending data
- Negative: Left-skewed (long tail left)
- Zero: Symmetric (normal distribution)

**Quartiles & IQR**
- Q1: 25th percentile (1/4 way through)
- Q2: 50th percentile (median)
- Q3: 75th percentile (3/4 way through)
- IQR = Q3 - Q1 (middle 50% of data)

## 4️⃣ Detect Outliers - See Unusual Values

````python
# Find outliers using IQR method
outliers = bd.outliers()

print("""
OUTLIER DETECTION FORMULA (IQR Method):
╔════════════════════════════════════════════════════════════╗
║ Lower bound = Q1 - 1.5 × IQR                              ║
║ Upper bound = Q3 + 1.5 × IQR                              ║
║                                                            ║
║ Values OUTSIDE these bounds = OUTLIERS                    ║
║ IQR = Q3 - Q1 (middle 50% of data)                        ║
╚════════════════════════════════════════════════════════════╝

Why 1.5 × IQR?
→ Statistical standard based on normal distribution
→ Catches ~0.7% of normal data as outliers
→ Good balance between sensitivity and specificity
""")
````

---

## 5️⃣ Normality Testing - Is Data Normal?

````python
# Test if data is normally distributed
normality = bd.normality_test()

print("""
SHAPIRO-WILK TEST - Is Data Normal?
╔════════════════════════════════════════════════════════════╗
║ H0 (Null Hypothesis): Data IS normally distributed         ║
║                                                            ║
║ If p-value > 0.05: FAIL to reject H0 → Likely NORMAL     ║
║ If p-value < 0.05: REJECT H0 → Likely NOT NORMAL         ║
╚════════════════════════════════════════════════════════════╝

Interpretation for SPENDING:
• p-value should be < 0.05 (not normal)
• Because data is SKEWED (right tail)
• Real-world data often isn't normal!
""")
````

## 6️⃣ Analyze Relationships - Correlations

````python
# Analyze correlations between columns
corr_matrix = bd.correlations(method='pearson')

print(f"\n{corr_matrix}")
````

````python
print("""
CORRELATION FORMULA (Pearson)
╔════════════════════════════════════════════════════════════╗
║ r = Σ((x - μₓ)(y - μᵧ)) /                                ║
║     √(Σ(x - μₓ)² × Σ(y - μᵧ)²)                          ║
║                                                            ║
║ Range: -1 to +1                                           ║
║ r = +1: Perfect positive relationship                     ║
║ r = 0: No linear relationship                            ║
║ r = -1: Perfect negative relationship                     ║
╚════════════════════════════════════════════════════════════╝

Interpretation:
• 0.7 to 1.0 → Strong correlation
• 0.4 to 0.7 → Moderate correlation
• 0.0 to 0.4 → Weak correlation

⚠️ Important: CORRELATION ≠ CAUSATION
Two variables can correlate without one causing the other!
""")
````

## 8️⃣ Custom Analysis - Compare Groups

````python
# Analyze spending by lunch type (categorical)
import polars as pl

df_pl = bl.load_dataset('school_cafeteria')

# Group by lunch type
grouped = df_pl.group_by('lunch_type').agg([
    pl.col('spending').mean().alias('Avg_Spending'),
    pl.col('spending').median().alias('Median_Spending'),
    pl.col('spending').std().alias('Std_Dev'),
    pl.col('satisfaction').mean().alias('Avg_Satisfaction'),
    pl.col('*').count().alias('Count')
])

print("\n📊 SPENDING BY LUNCH TYPE")
print(grouped.to_pandas().to_string(index=False))

print("""
INSIGHTS:
• Which lunch type has highest spending?
• Which has highest satisfaction?
• Is high spending = high satisfaction?
• Why might pizza have different patterns?
""")
````

````python
# Visualize spending by lunch type
df_native = df_pl.to_native()
if hasattr(df_native, 'to_pandas'):
    df_native = df_native.to_pandas()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Box plot
df_native.boxplot(column='spending', by='lunch_type', ax=axes[0])
axes[0].set_title('Spending Distribution by Lunch Type')
axes[0].set_xlabel('Lunch Type')
axes[0].set_ylabel('Spending ($)')

# Scatter plot
for lunch_type in df_native['lunch_type'].unique():
    data = df_native[df_native['lunch_type'] == lunch_type]
    axes[1].scatter(data['spending'], data['satisfaction'], label=lunch_type, s=100, alpha=0.6)

axes[1].set_xlabel('Spending ($)')
axes[1].set_ylabel('Satisfaction (1-10)')
axes[1].set_title('Spending vs Satisfaction by Lunch Type')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
````

## Summary: What You've Learned

### ✅ Key Concepts

**1. Descriptive Statistics**
- Mean (average), Median (middle), Mode (most common)
- Std Dev (spread), Quartiles (divisions)
- Skewness (asymmetry), Kurtosis (tail heaviness)

**2. Distributions**
- Normal (bell curve) vs Skewed (asymmetric)
- Bimodal (two peaks) vs Unimodal (one peak)
- Real data often ISN'T perfectly normal

**3. Visualizations**
- Histogram: Shows frequency distribution
- Boxplot: Shows quartiles and outliers
- Violin: Shows full density distribution
- Heatmap: Shows correlations

**4. Outlier Detection**
- IQR method: Values beyond 1.5 × IQR
- Visual inspection: Use boxplots
- Domain knowledge: What makes sense in context

**5. Hypothesis Testing**
- Shapiro-Wilk: Test for normality
- p-value < 0.05: Likely NOT that hypothesis
- Important for choosing appropriate statistical tests

**6. Correlations**
- Pearson r: Linear relationship strength
- Range: -1 to +1
- Correlation ≠ Causation (critical!)
